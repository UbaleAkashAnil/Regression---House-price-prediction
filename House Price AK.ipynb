```python
# Creating list of Columns having Continuous data

Column_with_continuous_data = ['LotFrontage','LotArea','YearBuilt','YearRemodAdd','MasVnrArea','BsmtFinSF1',
                                'BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','GrLivArea','GarageYrBlt',
                               'GarageArea','WoodDeckSF','OpenPorchSF','MoSold']
```

```python
# Capping Of Outliers

def Outliers_Removal(Column_with_continuous_data):
    Q1=df1[Column_with_continuous_data].quantile(0.25)
    Q3=df1[Column_with_continuous_data].quantile(0.75)
    IQR=Q3-Q1
    LW= Q1-1.5*IQR
    RW= Q3+1.5*IQR
    df1[Column_with_continuous_data]=np.clip(df1[Column_with_continuous_data],LW,RW)
    
```

```python
for var in Column_with_continuous_data:
    Outliers_Removal(var)
```

```python
plt.figure(figsize=(20,30))

for i , variable in enumerate (Column_with_continuous_data):
    plt.subplot(10,4,i+1)
    plt.boxplot(df1[variable],whis=1.5)
    plt.tight_layout()
    plt.title(variable)
    
plt.show()
```

```python
# HAndel ccat VAlues

for cols in df1.columns:
    if df1[cols].dtype=='object':
        from sklearn.preprocessing import LabelEncoder
        le=LabelEncoder()
        df1[cols]=le.fit_transform(df1[cols])
```

## Data is Ready to learn from algorithm

```python
x = df1.drop(['Property_Sale_Price','Id'],axis=1)

y = df1[['Property_Sale_Price']]
```

# Data Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(x)
x_scaled=scaler.transform(x)
```

```python
## Training Testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20)
```

```python
print('x_train',x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

## Import The Algorithm
### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lg=LinearRegression()
lg
```

```python
lg.fit(x_train,y_train)
```

```python
print('Intercept:',lg.intercept_)
print('Coefficient:',lg.coef_)
```

```python
pred=lg.predict(x_test)
pred
```

## RSquare Score

```python
from sklearn.metrics import r2_score
y_test_pred=lg.predict(x_test)
y_train_pred=lg.predict(x_train)
a=r2_score(y_train,y_train_pred)
b=r2_score(y_test,y_test_pred)
print('traninig={} testing={}'.format(a,b))
```

```python
import numpy as np
from sklearn import metrics
```

```python
meanAbErr = metrics.mean_absolute_error(y_test, pred)
meanSqErr = metrics.mean_squared_error(y_test,pred)
#rootMeanSqErr = np.sqrt(mlr.mean_squared_error(y_test, y_pred_mlr))

print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
#print('Root Mean Square Error:', rootMeanSqErr)
```
